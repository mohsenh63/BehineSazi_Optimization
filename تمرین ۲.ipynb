تعریف تابع

```python
def f(x1, x2):
#     tavan  = (x1 - mt.pi) ** 2 - (x2 - mt.pi) ** 2
    f = -20 * mt.exp((-0.2 * mt.sqrt(x1 ** 2 + x2 ** 2) / 2)) - mt.exp(1/2*(mt.cos(2 * mt.pi * x1) + mt.cos(2 * mt.pi *x2))) + 20 + mt.exp(1) 
    
    return f
```

رسم سطح تراز

```python
import math as mt
import numpy as np
import matplotlib.pyplot as plt
import itertools

x = np.arange(-5, 5, 0.025)
y = np.arange(-5, 5, 0.025)
X, Y = np.meshgrid(x, y)
Z = np.zeros(X.shape)
for i, j in itertools.product(range(len(X)), range(len(X))):
        Z[i][j] =  f(X[i][j], Y[i][j])
onvan = r"$-20e^{\left(-0.2 \sqrt{\frac{x_1^2 + x_2^2}{2}}\right)} - e^{\left(\frac{1}{2}\left[cos(2\pi x_1) + cos(2\pi x_2)\right]\right)} + 20 + e^1$"        
fig, ax = plt.subplots(figsize=(10, 10))
pc = plt.contour(X, Y, Z)
ax.clabel(pc, fontsize=9, inline=True)
plt.xlabel( ' x1 ' )
plt.ylabel( ' x2 ' )
plt.title(onvan, fontsize = 20)

plt.show()
```
